```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/eda_cleaned_data.csv')
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
fig, ax = plt.subplots(figsize=(20,8))
g = sns.barplot(data=df, x = df.vehicle_type_code1.value_counts().index, y=df.vehicle_type_code1.value_counts(), palette= 'crest', ci=None)
g.set_xticklabels(g.get_xticklabels(), rotation=45, ha='right', fontsize=14)

plt.xlabel("Vehicle Type", fontsize=20)
plt.ylabel("Amount", fontsize=20)
plt.title("Frequency of Vehicles Involved", fontsize=32)
plt.tight_layout
plt.show()
```
